#### Let’s say you want to get all the beds from IKEA together with the product details to perform some analysis. 
Here is the request that my browser caught from the URL (https://www.ikea.com/au/en/cat/beds-bm003/).

```python
import pandas as pd
import requests
```

```python
json = requests.get(
'https://sik.search.blue.cdtapps.com/us/en/product-list-page?category=bm003&size=714&c=lf&v=20211124&sort=RELEVANCE&store=411&zip=20902&sessionId=571908c6-de46-4ca7-b136-3bfaa9dba917').json()['productListPage']['productWindow']
```

```python
df = pd.DataFrame.from_dict(json)
df
```

```python
# Export output to excel for analysis
df.to_csv('C:/Users/Joy/Desktop/data science/webscraping1.csv')
```
